```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
from scipy.stats import linregress as lr
```

```python
df = pd.read_csv('demo.csv')

df
```

```python
fig, ax = plt.subplots(1, 4, figsize = (20, 5))
for i in range(4):
    ax[i].scatter(df[f'x{i+1}'], df['y'])
```

```python
for i in range(4):
    _,_,r,p,_ = lr(df[f'x{i+1}'], df['y'])
    print(f'Between x{i+1} and y, R^2 = {r**2} \t p = {p}')
```

```python
X = df[['x1', 'x2']]
y = df[['y']]

model = lm.LinearRegression(fit_intercept=True)
model.fit(X,y)

model.score(X = X, y = y)
```

```python
fig, ax = plt.subplots(5, 5, figsize = (25,25))
for i in range(4):
    for j in range(4):
        ax[i,j].scatter(df[f'x{i+1}'], df[f'x{j+1}'])

for i in range(4):
    ax[i,4].scatter(df[f'x{i+1}'], df[f'y'])
    ax[4,i].scatter(df[f'y'], df[f'x{i+1}'])

ax[4,4].scatter(df[f'y'], df[f'y'])
```
